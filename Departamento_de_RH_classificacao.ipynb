```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./departamento_de_rh_tratado.csv')
df.shape
```

```python
df
```

```python
df.columns
```

```python
X_cat = df[['NAME_POSICAO', 'ESTADO_RESIDE', 'Idade','RACA', 'SITUACAO_PROFISSIONAL', 'DEPARTAMENTO', 'FONTE_RECRUTAMENTO', 'PONTUACAO_PERFORMACE',
       'HISPANICO/LATINO', 'SEXO', 'GERENTE_PESSOA']]
X_cat
```

```python
from sklearn.preprocessing import OneHotEncoder
onehotencoder = OneHotEncoder()
X_cat = onehotencoder.fit_transform(X_cat).toarray()
```

```python
X_cat.shape
```

```python
X_cat = pd.DataFrame(X_cat)
```

```python
X_cat
```

```python
X_numerical = df[['CASADO', 'ESTADO_CIVIL'
       , 'DIVERSIDADE', 'SALARIO', 'POSICAO',  'PONTUACAO_SATISFACAO','ENGAJAMENTO',
       'N_PROJETOS', 'N_ATRASOS', 'N_AUSENTE']]
X_numerical
```

```python
X_all = pd.concat([X_cat, X_numerical], axis = 1)
X_all
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X = scaler.fit_transform(X_all)
```

```python
X
```

```python
y = df['DEMITIDO']
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```

```python
X_train.shape, y_train.shape
```

```python
X_test.shape, y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
logistic =  LogisticRegression()
logistic.fit(X_train, y_train)
```

```python
y_pred = logistic.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True);
```

```python
import pickle
```

```python
with open('variaveis_modelo.pkl', 'wb') as f:
  pickle.dump([scaler, onehotencoder, logistic], f)
```
